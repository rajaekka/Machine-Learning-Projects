# Clustering Plant Iris Using Principle Componant Analysis


* Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

* Load Dataset

```python
from sklearn.datasets import load_iris
df=load_iris()
```

* Dataset Segregation

```python
X=df.data
y=df.target
names=df.target_names
```

* Fitting The PCA Clustering to the datasetn with n=2 

```python
from sklearn.decomposition import PCA
model=PCA(n_components=2)  
# number of componants  to keep is 2

y_means=model.fit(X).transform(X)
```

* Varience Percentage

```python
plt.figure()
colors=['m','k','c']

for color,i ,target_name in zip(colors,[0,1,2],names):
    plt.scatter(y_means[y==i,0],y_means[y==i,1],color=color,lw=2,label=target_name)
    
plt.title('Iris Clusterring')
plt.show()
```
